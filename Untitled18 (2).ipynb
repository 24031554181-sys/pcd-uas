```python
import os
import shutil
import random
import numpy as np
import tensorflow as tf

random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)



INPUT_DIR = "C:\pcd uas\data"
OUTPUT_DIR = "C:\pcd uas\output"
```

preprosessing


```python
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

if os.path.exists(OUTPUT_DIR):
    shutil.rmtree(OUTPUT_DIR)

for split in ["train", "val", "test"]:
    for category in os.listdir(INPUT_DIR):
        path = os.path.join(OUTPUT_DIR, split, category)
        os.makedirs(path, exist_ok=True)

for category in os.listdir(INPUT_DIR):
    category_path = os.path.join(INPUT_DIR, category)

    if not os.path.isdir(category_path):
        continue

    files = sorted(os.listdir(category_path))  

    total = len(files)
    train_end = int(train_ratio * total)
    val_end = int((train_ratio + val_ratio) * total)

    train_files = files[:train_end]
    val_files = files[train_end:val_end]
    test_files = files[val_end:]

    for f in train_files:
        shutil.copy(
            os.path.join(category_path, f),
            os.path.join(OUTPUT_DIR, "train", category, f)
        )

    for f in val_files:
        shutil.copy(
            os.path.join(category_path, f),
            os.path.join(OUTPUT_DIR, "val", category, f)
        )

    for f in test_files:
        shutil.copy(
            os.path.join(category_path, f),
            os.path.join(OUTPUT_DIR, "test", category, f)
        )
```

```python
TRAIN_DIR = os.path.join(OUTPUT_DIR, "train")
VAL_DIR   = os.path.join(OUTPUT_DIR, "val")
TEST_DIR  = os.path.join(OUTPUT_DIR, "test")

emotion_classes = sorted([
    d for d in os.listdir(TRAIN_DIR)
    if os.path.isdir(os.path.join(TRAIN_DIR, d))
])
print("Emotion Classes:", emotion_classes)
```

Augmentasi


```python
IMG_SIZE = 128      #
BATCH_SIZE = 32     #

# 1. Inisialisasi Detektor Wajah
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 2. Buat Fungsi Preprocessing Khusus Keras Generator
def remove_background_callback(img):
    # Keras melemparkan gambar berupa array float32 dengan range 0-255 (RGB)
    # Konversi ke uint8 dan Grayscale untuk OpenCV Haar Cascade
    img_uint8 = img.astype(np.uint8)
    gray = cv2.cvtColor(img_uint8, cv2.COLOR_RGB2GRAY)
    
    # Deteksi lokasi wajah
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    if len(faces) > 0:
        (x, y, w, h) = faces[0]
        # Potong wajah dari gambar asli
        cropped_img = img[y:y+h, x:x+w]
        # Resize kembali ke ukuran target awal (128x128) agar dimensinya tidak error saat dikumpulkan jadi batch
        img_processed = cv2.resize(cropped_img, (img.shape[1], img.shape[0]))
        return img_processed
        
    return img

# 3. Masukkan fungsi di atas ke dalam generator
train_datagen = ImageDataGenerator(
    rescale=1./255,                     #
    rotation_range=15,                  #
    zoom_range=0.2,                     #
    width_shift_range=0.1,              #
    height_shift_range=0.1,             #
    horizontal_flip=True,               #
    preprocessing_function=remove_background_callback # <--- FITUR BARU
)

val_datagen = ImageDataGenerator(
    rescale=1./255,                     #
    preprocessing_function=remove_background_callback # <--- FITUR BARU
)

test_datagen = ImageDataGenerator(
    rescale=1./255,                     #
    preprocessing_function=remove_background_callback # <--- FITUR BARU
)
```

data train


```python
train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="rgb",
    batch_size=BATCH_SIZE,
    shuffle=True,
    classes=emotion_classes,
    class_mode="sparse"
)

val_generator = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="rgb",
    batch_size=BATCH_SIZE,
    shuffle=False,
    classes=emotion_classes,
    class_mode="sparse"
)

test_generator = test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="rgb",
    batch_size=BATCH_SIZE,
    shuffle=False,
    classes=emotion_classes,
    class_mode="sparse"
)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
import tensorflow as tf

base_model = MobileNetV2(
    weights="imagenet",
    include_top=False,
    input_shape=(128, 128, 3)
)

base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)

x = Dense(256, activation="relu")(x)
x = Dropout(0.5)(x)

outputs = Dense(NUM_CLASSES, activation="softmax")(x)

model = Model(inputs=base_model.input, outputs=outputs)

model.compile(
    optimizer=tf.keras.optimizers.AdamW(learning_rate=1e-4),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
import joblib
from tqdm import tqdm
from skimage.feature import hog, local_binary_pattern
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.svm import  SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python

emotion_classes = ['Angry', 'Happy', 'Sad']

label_map = {cls: idx for idx, cls in enumerate(emotion_classes)}

IMG_SIZE = 128  

print("Label Map sukses dibuat:", label_map)
```

```python
# 1. Inisialisasi Detektor Wajah bawaan OpenCV (Taruh di luar fungsi)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def ekstrak_fitur_dataset(folder_root):
    X_lokal = []
    y_lokal = []
    
    for class_name, label in label_map.items():
        folder = os.path.join(folder_root, class_name)
        if not os.path.exists(folder):
            continue
            
        images = os.listdir(folder)
        for img_name in tqdm(images, desc=f"Memproses {class_name} ({os.path.basename(folder_root)})"):
            img_path = os.path.join(folder, img_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue
                
            # =======================================================================
            # 🟢 PROSES REMOVE BACKGROUND (OTOMATIS FOKUS FOREGROUND / WAJAH)
            # =======================================================================
            # Mendeteksi lokasi wajah di dalam gambar
            faces = face_cascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
            
            if len(faces) > 0:
                # Jika wajah ditemukan, ambil wajah pertama lalu potong (Crop)
                # Langkah ini otomatis membuang 100% background di luar kotak wajah
                (x, y, w, h) = faces[0]
                img_processed = img[y:y+h, x:x+w]
            else:
                # Jika deteksi gagal (karena foto terlalu gelap/miring), 
                # gunakan gambar asli agar program tidak crash
                img_processed = img
            
            # Resize gambar hasil crop wajah ke ukuran standar (128x128)
            img_resized = cv2.resize(img_processed, (IMG_SIZE, IMG_SIZE))

            # 2. Ekstraksi Fitur HOG dari wajah yang sudah bersih dari background
            hog_features = hog(
                img_resized, orientations=9, pixels_per_cell=(8, 8),
                cells_per_block=(2, 2), block_norm='L2-Hys'
            )

            X_lokal.append(hog_features)
            y_lokal.append(label)
            
    return np.array(X_lokal), np.array(y_lokal)

# --- Jalankan ulang ekstraksi datanya ---
print("--- Ekstraksi Data TRAIN (Bersih Background) ---")
X_train_raw, y_train = ekstrak_fitur_dataset(r"C:\pcd uas\output\train")

print("\n--- Ekstraksi Data VALIDATION (Bersih Background) ---")
X_val_raw, y_val = ekstrak_fitur_dataset(r"C:\pcd uas\output\val")

print("\n--- Ekstraksi Data TEST (Bersih Background) ---")
X_test_raw, y_test = ekstrak_fitur_dataset(r"C:\pcd uas\output\test")

print(f"\nSelesai! Train: {X_train_raw.shape}, Val: {X_val_raw.shape}, Test: {X_test_raw.shape}")
```
